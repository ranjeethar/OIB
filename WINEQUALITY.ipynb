```python
# import librares
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
#import dataset
data = pd.read_csv('WineQT.csv')
```

```python
data.info()
```

```python
#printing rows of dataset
print("First few rows of the dataset:")
print(data.head())
```

```python
#calculating random forest accuracy
H = data.drop('quality', axis=1)
p = data['quality']
H_train, H_test, p_train, p_test = train_test_split(H, p, test_size=0.2, random_state=12)
rf_classifier = RandomForestClassifier()
rf_classifier.fit(H_train, p_train)
rf_predictions = rf_classifier.predict(H_test)
rf_accuracy = accuracy_score(p_test, rf_predictions)
print("Random Forest Accuracy:", rf_accuracy)
```

```python
#Stocastic Gradient Descent Classifier Accuracy
sgd_classifier = SGDClassifier()
sgd_classifier.fit(H_train, p_train)
sgd_predictions = sgd_classifier.predict(H_test)
sgd_accuracy = accuracy_score(p_test, sgd_predictions)
print("Stocastic Gradient Descent Classifier Accuracy:", sgd_accuracy)
```

```python
#svc classifier
svc_classifier = SVC()
svc_classifier.fit(H_train, p_train)
svc_predictions = svc_classifier.predict(H_test)
svc_accuracy = accuracy_score(p_test, svc_predictions)
print("Support Vector Classifier Accuracy:", svc_accuracy)
```

```python
#finding density
density_array = data['density'].values
alcohol_array = data['alcohol'].values
```

```python
#finding mean and standard deviation
mean_density = np.mean(density_array)
mean_alcohol = np.mean(alcohol_array)
std_density = np.std(density_array)
std_alcohol = np.std(alcohol_array)
print("\nStatistics for Density:")
print("Mean Density:", mean_density)
print("Standard Deviation Density:", std_density)
print("\nStatistics for Alcohol:")
print("Mean Alcohol:", mean_alcohol)
print("Standard Deviation Alcohol:", std_alcohol)
```

```python
#data visualisatio
# heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
#countplot visulaisation
sns.countplot(data['quality'], palette='flare')
plt.title('Distribution of Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
#scatterplot
plt.figure(figsize=(8, 6))
plt.scatter(data['alcohol'], data['density'], alpha=0.2, color='green')
plt.title('Alcohol Content vs. Density')
plt.xlabel('Alcohol Content')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
avg_quality = data.groupby('quality')['quality'].count()
# Plotting the bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=avg_quality.index, y=avg_quality.values, palette='viridis')
plt.title('Average Quality of Wine for Each Quality Level')
plt.xlabel('Quality Level')
plt.ylabel('Average Quality')
plt.show()
```

```python
# Plotting the histogram
plt.figure(figsize=(8, 6))
plt.hist(data['quality'], bins=range(1, 11), edgecolor='black', alpha=0.7)
plt.title('Distribution of Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.xticks(range(1, 11))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='quality', y='pH', data=data, palette='coolwarm')
plt.title('Boxplot of pH for Each Wine Quality Level')
plt.xlabel('Quality Level')
plt.ylabel('pH')
plt.show()
```
